## Define Example Inputs

### Subtask:
Create a list of example strings (e.g., potential ham messages and potential spam messages) that the user wants to test with the model.


**Reasoning**:
The subtask requires creating a Python list of example messages. This code block will initialize the `example_messages` list as requested.



```python
example_messages = [
    "Hey, how are you doing today?",
    "Meeting at 3 PM today.",
    "WINNER! You have won a free iPhone. Claim now!",
    "URGENT! Your account has been suspended. Click here to reactivate it.",
    "Can we catch up tomorrow?",
    "Free entry to a contest to win cash! Text YES to 8888."
]

print("Example messages list created successfully.")
```

**Reasoning**:
The next step is to preprocess the newly created `example_messages` list using the same cleaning and stemming steps applied to the original dataset. This involves removing special characters, converting to lowercase, tokenizing, removing stopwords, and stemming each word.



```python
preprocessed_example_messages = []
for message in example_messages:
    review = re.sub('[^a-zA-Z]', ' ', message)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = " ".join(review)
    preprocessed_example_messages.append(review)

print("Example messages preprocessed successfully.")
print("Preprocessed messages:", preprocessed_example_messages)
```

**Reasoning**:
The preprocessed example messages now need to be converted into a numerical feature representation (Bag of Words) using the `CountVectorizer` that was previously fitted on the training data. This ensures consistency in the feature space for prediction.



```python
X_example = cv.transform(preprocessed_example_messages).toarray()

print("Example messages vectorized successfully.")
print("Shape of vectorized example messages:", X_example.shape)
```

**Reasoning**:
Now that the example messages have been preprocessed and vectorized, the next logical step is to use the trained Multinomial Naive Bayes classifier to predict whether each message is 'ham' (0) or 'spam' (1).



```python
predictions = classifier.predict(X_example)

print("Predictions for example messages generated successfully.")
print("Predictions:", predictions)
```

**Reasoning**:
To provide a comprehensive output, the original message, its preprocessed form, and the corresponding prediction need to be displayed together. This will allow for easy interpretation of the model's performance on the example messages.



```python
print("\n--- Model Predictions on Example Messages ---")
for i, (original_msg, preprocessed_msg, prediction) in enumerate(zip(example_messages, preprocessed_example_messages, predictions)):
    label = "Spam" if prediction == 1 else "Ham"
    print(f"\nMessage {i+1}:")
    print(f"  Original: {original_msg}")
    print(f"  Preprocessed: {preprocessed_msg}")
    print(f"  Predicted Label: {label} ({prediction})")
```
